```python
import pandas as pd
from geopy.geocoders import Nominatim  # Make sure to install geopy: pip install geopy

# Sample DataFrame with latitude and longitude columns
df = pd.DataFrame(df)

# Create a geolocator
geolocator = Nominatim(user_agent="geo_locator")

# Loop through each row in the DataFrame
for index, row in df.iterrows():
    latitude = str(row['lat'])
    longitude = str(row['lng'])

    # Get location information
    location = geolocator.reverse(latitude + "," + longitude)
    address = location.raw.get('address', {})

    # Extract city and country
    city = address.get('city', '')
    country = address.get('country', '')

    # Update the DataFrame with city information
    df.at[index, 'City'] = city
    df.at[index, 'Country'] = country

    # Print the results
    print(f'Row {index + 1}:')
    #print('Latitude:', latitude)
    #print('Longitude:', longitude)
    #print('City:', city)
    #print('Country:', country)

# Display the updated DataFrame
print("Updated DataFrame:")
print(df)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim  # Make sure to install geopy: pip install geopy

# Sample DataFrame with a 'City' column
data_city = {'City': ['Paris', 'London', 'Berlin']}
df1 = pd.DataFrame(data_city)

# Create a geolocator
geolocator = Nominatim(user_agent="geo_locator")

# Loop through each row in the DataFrame
for index, row in df1.iterrows():
    city = row['City']

    # Get location information
    location = geolocator.geocode(city)

    # Update the DataFrame with latitude and longitude
    if location:
        df1.at[index, 'lat_updated'] = location.latitude
        df1.at[index, 'lon_updated'] = location.longitude
    else:
        print(f"Could not find coordinates for {city}")

# Display the updated DataFrame
print("Updated DataFrame:")
print(df1)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Hypothetical data - Replace this with your actual data
city_data = {
    'City': ['City A', 'City B', 'City C'],
    'Latitude': [40.7128, 34.0522, 41.8781],
    'Longitude': [-74.0060, -118.2437, -87.6298],
    'HotelCount': [50, 30, 70]
}

# Create a GeoDataFrame from the data
gdf = gpd.GeoDataFrame(city_data, geometry=gpd.points_from_xy(city_data['Longitude'], city_data['Latitude']))

# Plot the map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='lightgray', edgecolor='black', figsize=(10, 6))

# Plot hotels on top of the map
gdf.plot(ax=ax, color='red', marker='o', markersize=gdf['HotelCount'], alpha=0.7, legend=True)

# Add labels for each city
for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf['City']):
    ax.text(x, y, label, fontsize=8, ha='right')

# Add legend and title
plt.legend(['Hotels'])
plt.title('Hotel Distribution in Cities')

# Show the plot
plt.show()
```
